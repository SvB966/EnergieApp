```python
from IPython.display import display, HTML
import ipywidgets as widgets

display(HTML("""
<style>
    /* Algemene, consistente styling voor widgets, tekst, enz. (gebaseerd op 001_All_Types) */
    body, .widget-label, .widget-button, .widget-text, .widget-dropdown,
    .widget-html, .widget-select, .widget-checkbox, .widget-accordion,
    .widget-intprogress, .widget-progress {
        font-family: 'Roboto', sans-serif !important;
        font-size: 15px !important;
        color: #333333 !important;
    }
    .widget-text, .widget-dropdown {
        border-radius: 5px !important;
        border: 1px solid #ccc !important;
        padding: 6px 10px !important;
        margin: 3px 0px !important;
        box-sizing: border-box;
        background-color: #fff !important;
    }
    /* Knoppenstijl uit 001_All_Types, vergroot breedte om tekstafkapping te voorkomen */
    .widget-button {
        margin: 8px 5px !important;
        padding: 8px 16px !important;
        min-width: 200px !important; /* voorkom truncation */
        max-width: 240px !important; /* redelijk maximum, kan naar wens verder vergroot worden */
        height: auto !important;     /* laat de hoogte meegroeien als tekst op 2 regels moet */
        border: none !important;
        border-radius: 5px !important;
        background: linear-gradient(135deg, #0069d9, #004a9b) !important;
        color: #ffffff !important;
        font-weight: bold !important;
        transition: transform 0.2s ease, background 0.2s ease !important;
        text-align: center !important;
        white-space: normal !important; /* sta tekst op 2 regels toe indien nodig */
    }
    .widget-button:hover {
        transform: translateY(-1px) !important;
        background: linear-gradient(135deg, #007bff, #0056b3) !important;
    }
    .widget-button:active {
        transform: scale(0.98) !important;
    }
    .widget-button:disabled {
        opacity: 0.6 !important;
        cursor: not-allowed !important;
    }

    /* Placeholder-kleur voor widget-text */
    .widget-text::placeholder {
        color: rgba(150,150,150,0.7) !important;
    }

    /* Titels en subtitels (oorspronkelijk uit Start_UI) */
    .start-title {
        font-size: 28px;  /* iets groter */
        font-weight: bold;
        margin: 20px 0 10px 0;
        color: #0078D4;
        text-align: center;
    }
    .start-subtitle {
        font-size: 16px;
        margin-bottom: 25px;
        color: #333333;
        text-align: center;
    }
    /* Centraal blok met maximale breedte en schaduw */
    .vbox-container {
        display: flex;
        flex-direction: column;
        align-items: center;
        background-color: #FFFFFF;
        max-width: 600px;
        margin: 40px auto;
        padding: 30px;
        border-radius: 8px;
        box-shadow: 0 2px 5px rgba(0,0,0,0.15);
    }
    /* Logo-animatie behouden */
    .logo-container {
        text-align: center;
        margin-bottom: 20px;
    }
    .logo-container img {
        width: 150px;
        height: auto;
        animation: rotation 6s infinite linear;
    }
    @keyframes rotation {
        from { transform: rotate(0deg); }
        to { transform: rotate(360deg); }
    }
    /* Redirect-link en bedrijfsbeschrijving (oorspronkelijk uit Start_UI) */
    .redirect-link {
        font-size: 16px;
        color: #0078D4;
        text-decoration: none;
        margin-top: 20px;
        display: block;
        text-align: center;
    }
    .redirect-link:hover {
        text-decoration: underline;
    }
    .company-description {
        font-size: 14px;
        color: #333333;
        margin-top: 20px;
        text-align: center;
    }
</style>
"""))

# ----------------------------------------------------------------------------
# Output-widget voor JavaScript-redirect (Voila-compatibel)
# ----------------------------------------------------------------------------
redirect_out = widgets.Output()

def auto_redirect(url):
    """
    Voert een automatische redirect uit naar de gegeven URL.
    """
    with redirect_out:
        redirect_out.clear_output()
        display(HTML(f"<script>window.open('{url}', '_self');</script>"))

# ----------------------------------------------------------------------------
# Definitie van de knoppen
# ----------------------------------------------------------------------------
btn_energiemonitor      = widgets.Button(description="Energiemonitor",      icon='line-chart')
btn_dataexport          = widgets.Button(description="Export Data Analyse",  icon='table')
btn_vmned_dataexport    = widgets.Button(description="Export Data VMNED",    icon='file')
btn_factorupdate        = widgets.Button(description="Factorupdate",         icon='calculator')
btn_mvswitch            = widgets.Button(description="MV Switch",             icon='retweet')
btn_vervanging          = widgets.Button(description="Vervanging Tool",       icon='exchange')
btn_Storage_Method      = widgets.Button(description="Storage Method",        icon='clock-o')
btn_exit                = widgets.Button(description="Afsluiten",             icon='times')

# ----------------------------------------------------------------------------
# Callback-functies voor de knoppen
# ----------------------------------------------------------------------------
def on_energiemonitor_click(_):
    auto_redirect("http://127.0.0.1:8866")

def on_dataexport_click(_):
    auto_redirect("http://127.0.0.1:8867")

def on_vmned_dataexport_click(_):
    auto_redirect("http://127.0.0.1:8869")

def on_factorupdate_click(_):
    auto_redirect("http://127.0.0.1:8870")

def on_mvswitch_click(_):
    auto_redirect("http://127.0.0.1:8871")

def on_vervanging_click(_):
    auto_redirect("http://127.0.0.1:8872")

def on_Storage_Method_click(_):
    auto_redirect("http://127.0.0.1:8873")


def on_exit_click(_):
    display(HTML("<h3 style='text-align: center;'>Applicatie afgesloten.</h3>"))

btn_energiemonitor.on_click(on_energiemonitor_click)
btn_dataexport.on_click(on_dataexport_click)
btn_vmned_dataexport.on_click(on_vmned_dataexport_click)
btn_factorupdate.on_click(on_factorupdate_click)
btn_mvswitch.on_click(on_mvswitch_click)
btn_vervanging.on_click(on_vervanging_click)
btn_Storage_Method.on_click(on_Storage_Method_click)
btn_exit.on_click(on_exit_click)

# ----------------------------------------------------------------------------
# Titel, subtitel, logo en bedrijfsbeschrijving
# ----------------------------------------------------------------------------
title               = widgets.HTML("<div class='start-title'>Energymonitor</div>")
subtitle            = widgets.HTML("<div class='start-subtitle'>Selecteer:</div>")
logo                = widgets.HTML("<div class='logo-container'><img src='Innax_logo.jpg' alt='Innax Logo'></div>")
company_description = widgets.HTML("<div class='company-description'>INNAX Meten B.V. | Afdeling Data Management</div>")


ui = widgets.VBox(
    [
        logo,
        title,
        subtitle,
        btn_energiemonitor,
        btn_dataexport,
        btn_vmned_dataexport,
        btn_factorupdate,
        btn_mvswitch,
        btn_vervanging,
        btn_Storage_Method,
        btn_exit,
        company_description,
        redirect_out
    ],
    layout=widgets.Layout(justify_content='center', align_items='center', width='100%')
)
ui.add_class("vbox-container")

display(ui)
```
